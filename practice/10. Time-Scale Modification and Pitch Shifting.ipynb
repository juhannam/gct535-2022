```python
import librosa
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
import pytsmod as tsm   # note that you should install pytsmod (i.e., pip install pytsmod)
```

# Resampling

```python
#x, sr = librosa.load("../audio/suzanne.wav") 
x, sr = librosa.load("../audio/095_Coffee_House_10s.wav") 

sr2 = 1.2*sr

y_re = librosa.resample(x, sr, sr2)

ipd.Audio(y_re, rate=sr)
```

# Time-Scale Modification

```python
x, sr = librosa.load("../audio/suzanne.wav") 
#x, sr = librosa.load("../audio/095_Coffee_House_10s.wav") 

alpha = 1.5
y_tsm1 = tsm.wsola(x, alpha)

ipd.Audio(y_tsm1, rate=sr)
```

# Pitch Shifting ( WSOLA + Resampling) 

```python
#x, sr = librosa.load("../audio/suzanne.wav") 
x, sr = librosa.load("../audio/095_Coffee_House_10s.wav") 

alpha = np.power(2, 2/12)
y_tsm1 = tsm.wsola(x, alpha)

y_ps1 = librosa.resample(y_tsm1, sr, sr/alpha)

ipd.Audio(y_ps1, rate=sr)
```
